```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
base_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
```

```python
x = layers.GlobalAveragePooling2D()(base_model.output)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.2)(x)
output = layers.Dense(4, activation='softmax')(x)
```

```python
model = models.Model(inputs=base_model.input, outputs=output)
```

```python
model.compile(optimizer=optimizers.Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
```

```python

validation_datagen = ImageDataGenerator(rescale=1.0 / 255)
```

```python
train_dir = "D:\Lux Data Analytics Bootcamp\Python\Animal Detection Model/train"
validation_dir = "D:\Lux Data Analytics Bootcamp\Python\Animal Detection Model/Validation"
test_dir = "D:\Lux Data Analytics Bootcamp\Python\Animal Detection Model/test"
```

```python
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
```

```python
validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
```

```python

train_generator = train_datagen.flow_from_directory(train_dir, target_size=(224, 224), 
                                                    batch_size=32, class_mode='categorical')
validation_generator = validation_datagen.flow_from_directory(validation_dir, target_size=(224, 224), 
                                                              batch_size=32, class_mode='categorical')
```

```python
model.fit(train_generator, epochs=10, validation_data = validation_generator)
```
